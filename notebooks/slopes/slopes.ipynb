# Eulers number e

A property of e is that the slope of e^x = e.  So here is plotted n^x and its slope (derivative).  When the two lines match the value of n is e.

To plot the exponential function f(x)=nx and its derivative (the slope), we use the rule:
$$\frac{d}{dx} n^x = \ln(n) \cdot n^x$$

Note that when n=e≈2.718, the slope is exactly equal to the function itself. Since your range is 2≤n≤3, you can observe the slope transition from being smaller than the function to being larger than the function as n crosses the value of e.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def plot_exponential_slope(n):
    # Define x range
    x = np.linspace(-2, 4, 500)
    
    # Calculate the function y = n^x
    y = n**x
    
    # Calculate the slope (derivative) y' = ln(n) * n^x
    slope = np.log(n) * (n**x)
    
    plt.figure(figsize=(10, 6))
    
    # Plot the original function
    plt.plot(x, y, label=f'$f(x) = {n:.2f}^x$', color='blue', linewidth=2)
    
    # Plot the slope
    plt.plot(x, slope, label=f"$f'(x) = \\ln({n:.2f}) \\cdot {n:.2f}^x$", 
             color='red', linestyle='--')
    
    # Aesthetics
    plt.title(f'Function $n^x$ vs. its Slope (Base $n = {n:.2f}$)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.grid(True, which='both', linestyle='--', alpha=0.5)
    plt.legend()
    
    # Set y-limit to keep the view stable
    plt.ylim(-1, 30)
    plt.show()

# Create the interactive slider widget
interact(plot_exponential_slope, 
         n=FloatSlider(min=2.0, max=3.0, step=0.01, value=2.0, description='Base n:'))
```
